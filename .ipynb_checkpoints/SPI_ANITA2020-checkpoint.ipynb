In this notebook, you will see a practical example with real INTEGRAL data. The dataset has been provided. I have also provided you with a number of functions that will load the data ready for use (see the 'spipy.py' file).

```python
import os
import numpy
from scipy.ndimage.filters import gaussian_filter
import spipy

CWD = os.getcwd()
```

Firstly, load the data using the 'get_data' function.

```python
data = spipy.get_data(CWD+'/Inputs/evts_det_spec.fits', 50)
```

The data is stored as counts per detector, per pointing, per energy bin (channel).
Next we must load in information about what the energy boundaries of each energy bin is:

```python
energy_boundaries = spipy.get_energybins(CWD+'/Inputs/energy_boundaries.fits')
print(energy_boundaries)
```

Here you see the edges of each energy bin in keV.

Let's see what the spectrum we have looks like. 
For example, the data obtained by detector 3 in pointing 4:

```python
spipy.plotspectrum_raw(3, 4, energy_boundaries, data, 'rawspec', 'Outputs')
```

import the fitted data

```python
fitted = spipy.get_fitspec(CWD+'/fitted/spectra_Crab.fits')
```

Fit a function (which you should have justified above) to the data

```python
fitparams = spipy.fit_noRMF(CWD+'/fitted/spectra_Crab.fits', energy_boundaries)
```

```python
spipy.plot_spectrumfit_noRMF(fitted, energy_boundaries, fitparams, 'Outputs', 'CRAB_specnoRMF', ext = 'png')
```

We can look at the energy redistribution function for INTEGRAL here. Use the functions that allow you to load and plot the response matrix function

```python
RMF = spipy.get_RMFmatrix(CWD+'/fitted/spectral_response.rmf.fits', log = True)
#Do not mess with the next 2 lines
sigma = 0.4
RMF = gaussian_filter(RMF, sigma)
```

```python
x = energy_boundaries
#	get the new energy boundaries for the RMF matrix
y = spipy.get_RMF_energyboundaries(CWD+'/fitted/spectral_response.rmf.fits')

#	plot the visualization of the RMF matrix
spipy.plot_RMFmatrix(RMF, x, y, 'Outputs', 'CRAB_RMFvis')
```
